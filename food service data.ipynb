```python
df = food.copy()
```

the columns YEARS_IN_BUSINESS and  NUMBER_OF_ROOMS has the same vlaue in the rows so it doesnt have any impact in analysis so we are removing them

```python
df=df.drop(['YEARS_IN_BUSINESS', 'NUMBER_OF_ROOMS'], axis=1)
```

removing the rows which are null from columns :['FOOD_BEVERAGE_PURCHASE', 'FOOD_PURCHASE', 'BEVERAGE_PURCHASE', 'TOTAL_CATEGORY_DRY', 'TOTAL_CATEGORY_FRESH', 'TOTAL_CATEGORY_FROZEN', 'TOTAL_CATEGORY_PROTEIN', '% Share of F&B Purchases from Company']

```python
columns_to_check = ['COUNTY_CODE','COUNTY_NAME',
    'FOOD_BEVERAGE_PURCHASE', 'FOOD_PURCHASE', 'BEVERAGE_PURCHASE', 
    'TOTAL_CATEGORY_DRY', 'TOTAL_CATEGORY_FRESH', 'TOTAL_CATEGORY_FROZEN', 
    'TOTAL_CATEGORY_PROTEIN', '% Share of F&B Purchases from Company'
]

# Drop rows where any of the specified columns have null values
df_cleaned = df.dropna(subset=columns_to_check)

```
